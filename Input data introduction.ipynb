It is a brief introduction about the input data in the given traffic dataset.

These data include:

1. initial osm file (.osm)
2. the visualization of the osm file (.png)
3. processed link file (link.csv)
4. processed node file (node.csv)
5. travel demand data (od.csv)

Here, we take San Francisco as an example to demonstrate the input data.

# 0. Specify the file path 

```python
import pandas as pd
import matplotlib.pyplot as plt
import osmnx as ox

cityNo = '01'
cityname = 'San_Francisco'
filename = '01_Input_data'
filepath = cityNo + '_' + cityname + '/' + filename + '/'
```

# 1. Network data
### Node attributes
- Node_ID: the unique identifier for each node
- Lon: longitude
- Lat: latitude
- Tract_Node: determine whether a node point is TAZ

### Node file display

```python
node_data = pd.read_csv(filepath+cityname+'_node.csv',nrows=10)
display(node_data)
```

### Link attributes
- Link_ID: unique identifier for each link
- From_Node_ID: the node ID of the starting point of this link
- To_Node_ID: the node ID of the ending point of this link
- Capacity: default link capacity (veh/h)
- Length: link length (meter)
- Free_Speed: default link free flow speed (km/h)
- Lanes: the number of lanes
- Link_Type: supported link types in [osm2gmns](https://osm2gmns.readthedocs.io/en/latest/functions.html#osm2gmns.osmnet.build_net.getNetFromFile) (1 for motorway, 2 for trunk, 3 for primary, 4 for secondary, 5 for tertiary)

### Link file display

```python
link_data = pd.read_csv(filepath+cityname+'_link.csv',nrows=10)
display(link_data)
```

### Network Graphical display

```python
osm_file = filepath + cityname+'.osm'
graph = ox.graph_from_xml(osm_file)
fig, ax = ox.plot_graph(graph)
plt.show()
```

# 2. Travel demand data

### OD file display
- O_ID: the node ID of the origin TAZ point
- D_ID: the node ID of the destination TAZ point
- OD_Number: the corresponding travel demand

```python
od_data = pd.read_csv(filepath+cityname+'_od.csv',nrows=10)
display(od_data)
```
